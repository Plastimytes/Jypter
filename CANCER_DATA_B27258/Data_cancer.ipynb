Marginal Probability

```python
bins_perimeter = [0, 5, 10, 15, 20, 25, 30]
labels_perimeter = ['0-5', '5-10', '10-15', '15-20', '20-25', '25-30']


## create a new column with the binned data
data_cancer['perimeter_bin'] = pd.cut(data_cancer['perimeter_mean'], bins = bins_perimeter, labels=labels_perimeter, right=False)

## Calculate the Marginal Probabilities
perimeter_probabilities = data_cancer['perimeter_mean'].value_counts(normalize=True).sort_index()
perimeter_probabilities
```

Probability of a tumor having a perimeter mean less than or equal to 20

```python

tot_tum =len(data_cancer)
tum_less_20=data_cancer[data_cancer["perimeter_mean"]<=20].shape[0]
prob_tum=tum_less_20/tot_tum
print(f"Probability of a tumor having a perimeter mean less than or equal to 20 is {prob_tum}")
```
